# 0.3 Fourier's Law: The Conduction Rate Equation

The most fundamental equations in heat transfer are the rate equations: one for each of the three modes (conduction, convection, and radiation).  These rate equations are boiled down to the basics: any complications in the material or environment are abstracted out.

<br>

We'll start with conduction rate equation, which is also known as Fourier's Law.

Consider the convection rate equation:

<br>

$$q = hA(T_s -T_\infty)$$

<br>

2. increasing the surface area $A$

# The Fin Equation

The fin equation is a version of the heat equation that is found by using a differential control volume (that is, by doing a derivation similar to what we did to find the heat equation).  If we assume a couple things about our system, we can get a manageable equation: 1D conduction, steady-state, constant k, no heat generation, no radiation, a uniform convection coefficient $h$, and a uniform radius.

<br>

<center>
<img src = https://github.com/MAugspurger/Heat_Transfer_ENGR_321/raw/main/2_Convection/Convection_images/skeptical.PNG width = 300>
</center>

<br>

Ok, ok, you are right that we're ignoring some potentially significant factors.  But you are wrong, in that you are colored chartreuse.  But do you really want to see the equation without those assumptions?

<br>

<center>
<img src = https://github.com/MAugspurger/Heat_Transfer_ENGR_321/raw/main/2_Convection/Convection_images/carry_on.PNG width = 180>
</center>

<br>

Alright then.  With these assumptions, we get this equation, where $x$ is the axial direction:

$$\frac{d^2 T}{d x^2}=\frac{h P}{k A_c}\left(T-T_{\infty}\right)$$

<br>

$P$ is the perimeter of the fin, $A_c$ is the cross-sectional area, $T_\infty$ is the bulk fluid temperature, and $T$ is the local fin temperature.

<br>

<center>
<img src = https://github.com/MAugspurger/Heat_Transfer_ENGR_321/raw/main/2_Convection/Convection_images/fin_equation.PNG width = 300>
</center>

<br>

What does this equation mean?  Let's assume that the local fin temperature is higher than $T_\infty$.  The equation tells us the curvature of the temperature field from its base to its tip.  That curve is always the same sign, but is more severe if these two terms are large:

<br>

$$ \frac{h}{k}    ~~and~~ \frac{P}{A_c} $$

## Solving the Fin Equation Analytically

If we use this simplified version of the fin equation:

<br>

$$\frac{d^2 T}{d x^2}=\frac{h P}{k A_c}\left(T-T_{\infty}\right)$$

<br>

the solution to the equation is:

<br>

$$\theta(x)=C_A e^{m x}+C_B e^{-m x}$$

<br>

where $m=\sqrt{hP/kA}$.  This general solution isn't too hard to deal with.  The problems come with the boundary conditions.  We often know the base temperature, so that's one BC.  But what's the second one?  If we choose a flux BC at the base, we're defining the parameter that we often want to find out.  But we also don't have much information (temperature or flux) about the tip of the fin:

<br>

* If the fin is very long, we can set the BC to $T(L) = T_\infty$, or to $dT/dx = 0$.  But that means our fin is longer than it should be.
* We can set the tip flux BC to a convection boundary ($h(T-T_\infty) = k \frac{dT}{dx}$).  But the solutions to that are very complicated.

<br>

Given the overall limitations of the fin equation, and the fact that oftentimes fins are used in banks of multiple fins, these analytical solutions are of limited use.  


## Solving the Fin Equation Numerically

So let's solve the fin equation numerically, and play with the solutions to see how it behaves.  Then in class we'll use a software package to deal with some more complex and realistic situations.

<br>

Happily, the code we developed for the 1D heat equation works here with some small adaptations.  Because we are "stepping" through space, we will need to set the two BCs at the base.  But then we can adjust our parameters to match some BCs at the tip of the fin.

<br>

We'll choose some random system parameters: a 10 cm long fin made of aluminum, with a diameter of 1 cm, surrounding by $20^{\circ} C$ air that is being blown across the fin ($h=50 \frac{W}{m^2K}$).  We'll set the base temperature to $100^{\circ} C$ (our first boundary condition):

```python
import numpy as np
import pandas as pd

# Fin parameters
L = 0.1         # m
r = 0.005       # m
P = 2*np.pi*r   # m
A = np.pi*r**2  # m2

# Thermal parameters
k = 180         # W/mK (aluminum)
h = 50          # W/m2K (forced convection with air)
T_inf = 25      # degrees C (air temperature)
T_base = 100            # degrees C

# Simulation parameters
n = 1000        # number of nodes
dx = L/(n-1)    # distance between nodes
```

\The biggest change we have to make from our earlier plane simulation is to change our left boundary condition from an adiabatic condition (no flux) to a flux BC.  This will determine the value of the temperature at the second node so that we can use our discretized equation (which requires knowing two node temperatures to the "left" of each calculated node).

<br>

Based on a rough estimate of the average temperature of the fin and a quick first law balance, we'll choose a flux of $10 ~\frac{kW}{m^2}$:

```python
# Second boundary conditions
flux_base = 5.0e4      # W/m2
print("Our first guess at a heat rate is", round(flux_base*A,2), "W.")
```

Now we can set up a Series to hold our temperature values, and then set our first temperature value.  The first point, then, has a temperature of $T_0 = 100^{\circ} C$.

```python
# Set up a Series to hold the temperature field
pos_array = np.linspace(0,L,n)
temps = pd.Series(index=pos_array, dtype=object)

# Set the first value
temps.iloc[0] = T_base
```

The second BC is a flux BC ($q^{{\prime}{\prime}} = -k\frac{dT}{dx}$), which determines the slope of the temperature field next to the boundary.  So since we know the boundary temperature value and the slope of the field, we should be able to set the temperature value for the second point.

<br>

To do this, we first rearrange Fourier's Law to recognize that:

<br>

$$\frac{dT}{dx} = -\frac{q^{{\prime}{\prime}}}{k}$$

<br>

In discrete form, this is:

<br>

$$\frac{T_1-T_0}{\Delta x} = -\frac{q^{{\prime}{\prime}}}{k}$$

<br>

And so we can calculate the second point $T_1$ as:

<br>

$$T_1 = T_0 -\frac{q^{{\prime}{\prime}} \Delta x}{k}$$

<br>

This just finds the change in temperature by using Fourier's Law and the slope of temperature field at the base.  Now we'll code that equation:

```python
temps.iloc[1] = T_base - (flux_base *dx)/k
temps.iloc[1]
```

As you'd expect, the first point is slightly cooler than $100^{\circ} C$.  

<br>

And now we can use our discretization of the second order derivative to solve for the temperature field.   Just as a reminder, the discretization of the second derivative with respect to x looks like this:

<br>

$$\frac{d^2 T}{d x^2} = \frac{T_{m+1} -2T_m + T_{m-1}}{\Delta x^2}$$

<br>

So now we turn our continous fin equation:

<br>

$$\frac{d^2 T}{d x^2}=\frac{h P}{k A_c}\left(T-T_{\infty}\right)$$

<br>

into a discretized fin equation:

<br>

$$\frac{T_{m+1} -2T_m + T_{m-1}}{\Delta x^2} = \frac{hP}{kA_c} (T_m - T_\infty)$$

<br>

Notice what this means: as $T_m$ changes along the fin, the *curvature* of the temperature field changes too!

<br>

Since we're "stepping" through space, we want an equation that solves for $T_{m+1}$ using the known values $T_m$ and $T_{m-1}$.  So we can rearrange like this:

<br>

$$T_{m+1} = \left( \frac{hP}{kA_c} (T_m - T_\infty)\Delta x^2 \right) +2T_m -T_{m-1} $$

<br>

Now we just code up that equation, and loop through the nodes:

```python
# Loop through nodes until we find a solution
def fin_sim():
    C = (h*P*dx**2)/(k*A)
    pos_array = np.linspace(0,L,n)
    temps = pd.Series(index=pos_array, dtype=object)
    temps.iloc[0] = T_base
    for i in range(n-1):
        if i == 0:
            temps.iloc[i+1] = temps.iloc[i] - (flux_base *dx)/k
        else:
                temps.iloc[i+1] = C*(temps.iloc[i] - T_inf) + 2*temps.iloc[i] - temps.iloc[i-1]

    # Plot the numerical and analytical solution
    temps.plot(label='Numerical Solution',xlabel='Axial Position on Fin (m)',
             ylabel='Temperature (C)', title='Temperature Field in Plate', legend=True);
    return temps
```

```python
# Restate the parameters to make them easier to change later
# Fin parameters
L = 0.1         # m
r = 0.005       # m
P = 2*np.pi*r   # m
A = np.pi*r**2  # m2

# Thermal parameters
k = 180         # W/mK (aluminum)
h = 50          # W/m2K (forced convection with air)
T_inf = 25      # degrees C (air temperature)
T_base = 100            # degrees C
flux_base = 5.0e4      # W/m2

# Simulation parameters
n = 1000        # number of nodes
dx = L/(n-1)    # distance between nodes

# And now run the simulation
temps = fin_sim()
```

<br>

Uh, yeah. Something's off there: the tip of the fin should probably not be hotter than either the base or the convective fluid, should it?  Look again at the fin equation:

<br>

$$\frac{d^2 T}{d x^2}=\frac{h P}{k A_c}\left(T-T_{\infty}\right)$$

<br>

All this equation does is define the curvature of the temperature field (that is, the second derivative).  And if $T(x)$ is roughly the same for the whole fin, as it is here, then that curvature will be roughly constant.

<br>

We need to increase the initial slope of the temperature field at the base of the fin so that the temperature does not start to go up at any point along the fin.  Let's increase the flux by an order of magnitude (to $5.0 \times 10^5$) and run again:



```python
# Set the flux BC to a higher value and rerun
flux_base = 5.0e5
temps = fin_sim()
```

As the local $T$ gets closer to the $T_{\infty}$, the field becomes almost linear (that is, no curvature), and when it goes below $T_{\infty}$, that curvature actually becomes negative.  As it happens, the analytical solutions to the fin equation often come in form of hyperbolic sines and cosines, which look like this:

<br>

<center>
<img src = https://github.com/MAugspurger/Heat_Transfer_ENGR_321/raw/main/2_Convection/Convection_images/cosh_sinh.PNG width = 250>
</center>

<br>


You can see that our temperature fields above look similar to these functions.  So our simulation is accurately reproducing the differential equation, but our results are not physically meaningful.

<br>

The problem is our boundary conditions.  We just guessed at a flux BC, and the initial slope of the temperature field makes all the difference in terms of the shape of the field.  So in the exercises below, makes adjustments to that flux until we can get physically meaningful results.

## Fin effectivness and efficiency

Ideally, a fin is just long enough to maximize heat transfer but not so long that material is wasted and convective flows are unnecessarily disrupted.  There are two common ways to measure fin performance: fin effectiveness and fin efficiency.

<br>

Fin effectiveness is defined as:

<br>

$$\varepsilon_f = \frac{heat~transfer~rate~through~fin}{heat~transfer~rate~through~the~base~with~no~fin}$$

<br>

Fin efficiency compares the actual heat transfer to an maximum heat transfer through that particular fin (that is, when the fin is all at the base temperature):

<br>

$$\eta_f = \frac{heat~transfer~rate~through~fin}{max~heat~transfer~rate~through~this~fin}$$

<br>

Fin effectiveness should be above at least 2.0 to justify the expense of the fin.  Fin efficiency ranges from 0 to 1.0.  In both cases, a higher value suggests a higher performance.

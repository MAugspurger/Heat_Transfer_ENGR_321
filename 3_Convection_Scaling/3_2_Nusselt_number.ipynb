##Or, how I learned to stop worrying and love nondimensional numbers


## Defining the Nusselt Number

You may remember that we defined the convection coefficient $h$ like this:

<br>

$$h \equiv \frac{-\left.k_f \frac{\partial T_f}{\partial x}\right|_{\text {wall }}}{\left(T_s-T_{\infty}\right)} \equiv \frac{\text { Actual heat flux }}{\text { Potential for heat flux }}$$

<br>

<center>
<img src = https://github.com/AugustanaPEA/ENGR_321/raw/main/Images/Scaling/last_week.PNG width = 250>
</center>

<br>

If you look at this carefully, you might wonder about the units.  The right-hand side is set up like a ratio, but while the numerator is a temperature slope ($dT/dx$), the denominator is a temperature difference.  This actually makes it a little hard to understand what $h$ is:  what does it mean to have $h=1$, or $h=1000$?  It's not clear.

<br>

To more clearly compare the *actual* heat flux with the *potential* heat flux, we need a better sense of how far apart $T_s$ and $T_\infty$ are.  For instance, look at this simple convective flow:

<br>

<center>
<img src = https://github.com/AugustanaPEA/ENGR_321/raw/main/Images/Scaling/length_char.PNG width = 250>
</center>

<br>

If $L_c$ is very small, then convection won't matter that much: conduction by itself would produce large fluxes, because the hot surface and cool fluid would already be near to each other.  But if $L_c$ is large, then you probably need some serious advection to get a cold fluid packets next to the wall.

<br>

But let's multiply both sides of our definition of $h$ by $L_c$.  To further clear things up, we'll divide both sides by $k_f$:

<br>

$$\frac{hL_c}{k_f} =\frac{\color{green}{-\left. k_f \frac{\partial T_f}{\partial x}\right|_{\text {wall }}}}  {\color{lightblue}{k_f \left(\frac{T_s-T_{\infty}}{L_c}\right)}}$$

<br>

Now we're comparing apples to apples!

<br>

<center>
<img src = https://github.com/AugustanaPEA/ENGR_321/raw/main/Images/Scaling/fluxes.PNG width = 250>
</center>

<br>

The numerator (in green) represents the actual heat flux from the wall, aided by whatever convective flows are pushing colder fluid near the hot wall.  The denominator (light blue) represents the heat flux we'd have if there were *no* convection, and instead there was a nice linear temperature field from the surface to $L_c$.  In other words, its a ratio that tells us *by what factor convection is increasing the heat transfer*.   And this wondrous factor is called the *Nusselt number*!

<br>

$$Nu = \frac{hL_c}{k_f} = \frac{\color{green}{\text { Flux with convection }}}{\color{lightblue}{\text { Flux if there were no convection}}}$$

<br>

<center>
<img src = https://github.com/AugustanaPEA/ENGR_321/raw/main/Images/Scaling/nusselt.PNG width = 250>
</center>

<br>

So the Nusselt number makes more intuitive sense than $h$.  But what *is* $L_c$?  It is called the characteristic length, and it varies by geometry.  The $L_c$ of a pipe flow might be the radius; the $L_c$ of a plate flow might be the length of the plate.  As with the Reynolds number, what matters is that when we compare similar scenarios, we use the same definition of $L_c$.
